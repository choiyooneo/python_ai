## 1-11 : 평균의 제곱과 제곱의 평균

```python
score1 = 10
score2 = 20
score3 = 30

n_student = 3

mean = (score1 + score2 + score3) / n_student
square_of_mean = mean**2 #평균의 제곱

mean_of_square = (score1**2 + score2**2 + score3**2)/n_student #제곱의 평균

print("square of mean: ", square_of_mean)
print("mean of square : ", mean_of_square)
```

##1-12 : 분산과 표준편차
image.png 

```python
score1 = 10
score2 = 20
score3 = 30

n_student = 3

score_mean = (score1 + score2 + score3) / n_student
square_of_mean = score_mean**2 #평균의 제곱

mean_of_square = (score1**2 + score2**2 + score3**2) / n_student #제곱의 평균

score_variance = mean_of_square - square_of_mean #분산 = 제곱의평균 - 평균의제곱
score_std = score_variance ** 0.5 # 분산에 root를 씌운다. 

print("mean : ", score_mean) #평균
print("variance : ", score_variance) #분산
print("standard deviation : ", score_std) #표준편차
```

### 1-87 : 분산과 표준편차(5)

```python
scores = [[10,15,20],[20,25,30],[30,35,40],[40,45,50]]

n_student = len(scores)
n_class = len(scores[0])

class_score_sums = list()
class_score_square_sums = list()

class_score_variances = list()
class_score_std = list()


# 0으로 세팅
for _ in range(n_class):
  class_score_sums.append(0)
  class_score_square_sums.append(0)


for student_scores in scores:
  for class_idx in range(n_class):
    class_score_sums[class_idx] += student_scores[class_idx]
    class_score_square_sums[class_idx] += student_scores[class_idx]**2

#print(class_score_sums, class_score_square_sums)


# 분산 

for class_idx in range(n_class):
  mos = class_score_square_sums[class_idx]/n_student
  som = (class_score_sums[class_idx]/n_student)**2

  variance = mos - som
  std = variance**0.5

  class_score_variances.append(variance)
  class_score_std.append(std)



print("variances:", class_score_variances)  

print("standard deviations:", class_score_std)
```
